```python
sns.lmplot(
    x = 'sample_size',
    y = 'value',
    hue = 'method',
    truncate = False,
    palette = {'camelot': 'black', 'tabula': 'grey'},
    markers=["x", "o"],
    data = output.query('feature == "procedure" and metric == "time"')\
        .groupby(['method', 'sample_size', 'replicate'])\
        .sum()\
        .reset_index(drop=False),
    height=5, 
    aspect=1.5/1
)

plt.tight_layout()
```
